```python
import numpy as np
import cv2
import os
import ndjson

# resme cevirip kaydetmek icindir
def vektordenResme(cizimler,boyut=256):
    canvas=np.zeros((boyut,boyut))
    for stroke in cizimler:
        x,y=stroke
        for i in range(len(x)-1):
            canvas =  cv2.line(canvas,(x[i],y[i]),(x[i+1],y[i+1]),255,3)
    return 255 -canvas

kayitYolu="C:/Users/ahmad/Desktop/25kategori1500Resim"        
dosyaAdresi = 'C:/Users/ahmad/Desktop/yenidatandjson'
kategorilerdizi = ["apple","banana","basketball","broccoli","broom", "bus", "camera", 
                   "carrot", "chair","circle","cloud","door","eye","flower","hand",
                   "ladder","parachute", "pear","pineapple","rainbow","square",
                   "star","strawberry","sun","triangle"]
for j in kategorilerdizi:
    acma_yolu = os.path.join(dosyaAdresi,f'full_simplified_{j}.ndjson')
    def cizimleri_yukle(dosyaAdresi,limit=1001):
        with open(dosyaAdresi) as f:
            data=ndjson.load(f)
            return data[:limit]

    data=cizimleri_yukle(acma_yolu)
    print(f"{len(data)} cizim yuklendi {j} datalari. ")
    os.makedirs(f"C:/Users/ahmad/Desktop/25kategori1500Resim/{j}")
    print(f"{j} olusturuldu. ")

    for i in range(len(data)-1):
        cizimler = data[i]['drawing']
        resim = vektordenResme(cizimler)
        kayit_yolu = os.path.join(kayitYolu,f'{j}', f'{j}{i}.png')
        cv2.imwrite(kayit_yolu, resim)
```
